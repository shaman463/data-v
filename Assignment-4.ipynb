```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
np.random.seed(42)
data = np.random.randint(1, 201, size=(100, 30))
df = pd.DataFrame(data)
print(df)
```

```python
df = df.mask((df >= 10) & (df <= 60), np.nan)

# row
print("Na count per row:")
print(df.isna().sum(axis=1))

#  column
print("\nNa count per column:")
print(df.isna().sum(axis=0))
```

```python
df = df.apply(lambda col: col.fillna(col.mean()), axis=0)
print(df)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.heatmap(df, cmap='Accent')
plt.title('Heatmap of the DataFrame')
plt.show()
```

```python
corr = df.corr()
num_cols_corr = (corr.abs() <= 0.7).any(axis=0).sum()
print("Number of columns having correlation <= 0.7:", num_cols_corr)
```
